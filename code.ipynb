```python
file_path = "story.txt"

# Open the file for reading
try:
    with open(file_path, "r") as file:
        text = file.read()  # Read the entire contents of the file
        print(text)  # Print the story to the console
except FileNotFoundError:
    print(f"The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import string

temp_text = ""
for letter in text:
    if letter not in string.punctuation:
        temp_text += letter
text = temp_text
print(text)
```

```python
# tokenize the text
import nltk
import string

nltk.download("punkt")  # Download the tokenizer data (if not already downloaded)
tokenized_text = nltk.word_tokenize(text)

print(tokenized_text)
```

```python
# mapping each unique word to an index
vocabulary_index = {}
index_vocabulary = {}
i = 0
for word in tokenized_text:
    if word not in vocabulary_index:
        vocabulary_index[word] = i
        index_vocabulary[i] = word
        i += 1
vocabulary_size = len(vocabulary_index)

print(vocabulary_index)
print(vocabulary_size)
```

```python
n = 4  # the value of n in n-gram
```

```python
# generate ngrams
from nltk.util import ngrams
import numpy as np

ngrams = np.array(list((ngrams(tokenized_text, n))))

# Print the ngrams
print(ngrams)

print(ngrams.shape)
```

```python
from gensim.models import KeyedVectors

# Path to the downloaded Word2Vec binary file
model_path = "D:\deep learning\GoogleNews-vectors-negative300.bin.gz"

# Load the Word2Vec model
word2vec_model = KeyedVectors.load_word2vec_format(model_path, binary=True)
```

```python
# Test the loaded model
similar_words = word2vec_model.most_similar("king")
print("Words most similar to 'king':")
for word, score in similar_words:
    print(f"{word}: {score}")
```

```python
import numpy as np

# separate into x-set and y-set. The x-set will contain word embeddings of first n-1 words of each ngram. The y-set will contain the
# one-hot encodings of the last word of each ngram
x = []
y = []
for ngram in ngrams:
    single_example = []
    for i in range(n - 1):
        if word2vec_model.__contains__(ngram[i]):
            single_example.append(word2vec_model[ngram[i]].flatten())
        # elif ngram[i] == ".":
        #     single_example.insert(1, [0 for j in range(300)])
        else:
            single_example.insert(0, [0 for j in range(300)])

    y.append(
        vocabulary_index[ngram[n - 1]]
    )  # convert word to its corresponding integer as defined by vocabulary_index
    x.append(single_example)

x = np.array(x)  # convert x into numpy array

# convert each integer in y to one-hot encoding
y = np.eye(vocabulary_size)[y]

print(x.shape)
print(y.shape)
```

```python
print(x)
```

```python
print(y)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Bidirectional, Dense, Input
from keras.initializers import RandomUniform


model = Sequential(
    [
        Input(shape=(x.shape[1], x.shape[2])),
        Bidirectional(LSTM(units=10, activation="linear")),
        Dense(
            units=vocabulary_size,
            kernel_initializer=RandomUniform(minval=-1, maxval=1),
            activation="softmax",
        ),
    ]
)

model.compile(loss="categorical_crossentropy", optimizer="adam")
```

```python
model.fit(x, y, batch_size=32, epochs=100)
```

```python
def predict_next_word(partial_sentence):
    tokens = partial_sentence.split(" ")
    x = []
    for i in range(len(tokens) - n + 1, len(tokens)):
        if word2vec_model.__contains__(tokens[i]):
            x.append(word2vec_model[tokens[i]].flatten())
        else:
            x.append([0 for j in range(300)])

    x = np.expand_dims(x, axis=0)

    prediction = model.predict(x)
    return index_vocabulary[np.argmax(prediction)]


print(predict_next_word("you can"))
```

```python
def generate_sentence(seed, number_of_words_to_generate):
    sentence = seed
    for i in range(number_of_words_to_generate):
        next_word = predict_next_word(sentence)
        sentence = sentence + " " + next_word
    return sentence


print(
    generate_sentence(
        "the last thing",
        number_of_words_to_generate=10,
    )
)
```
